```python
team_city = {
    'ATL': 'Atlanta',
    'CHI': 'Chicago',
    'CON': 'Uncasville',
    'DAL': 'Arlington',
    'IND': 'Indianapolis',
    'LAS': 'Los Angeles',
    'LVA': 'Las Vegas',
    'MIN': 'Minneapolis',
    'NYL': 'Brooklyn',
    'PHO': 'Phoenix',
    'SEA': 'Seattle',
    'WAS': 'Washington DC'
}
```

```python
travel_tier = {}

teams = ['ATL', 'CHI', 'CON', 'DAL', 'IND', 'LAS', 'LVA', 'MIN', 'NYL', 'PHO', 'SEA', 'WAS']

# basic tier logic
for from_team in teams:
    for to_team in teams:
        if from_team == to_team:
            tier = 0
        elif {from_team, to_team} <= {'NYL', 'WAS', 'CON'}:
            tier = 1
        elif {from_team, to_team} <= {'CHI', 'IND', 'MIN'}:
            tier = 1
        elif {from_team, to_team} <= {'LAS', 'LVA', 'PHO'}:
            tier = 1
        elif (from_team in {'NYL', 'CON', 'WAS'} and to_team in {'CHI', 'IND', 'MIN'}) or \
             (from_team in {'CHI', 'IND', 'MIN'} and to_team in {'NYL', 'CON', 'WAS'}):
            tier = 2
        elif (from_team in {'ATL', 'DAL'} and to_team in {'CHI', 'IND', 'MIN'}) or \
             (to_team in {'ATL', 'DAL'} and from_team in {'CHI', 'IND', 'MIN'}):
            tier = 2
        elif (from_team in {'DAL', 'ATL'} and to_team in {'WAS', 'NYL', 'CON'}) or \
             (to_team in {'DAL', 'ATL'} and from_team in {'WAS', 'NYL', 'CON'}):
            tier = 2
        elif (from_team in {'SEA', 'PHO', 'LAS', 'LVA'} and to_team in {'ATL', 'WAS', 'NYL', 'CON', 'CHI', 'IND', 'MIN', 'DAL'}) or \
             (to_team in {'SEA', 'PHO', 'LAS', 'LVA'} and from_team in {'ATL', 'WAS', 'NYL', 'CON', 'CHI', 'IND', 'MIN', 'DAL'}):
            tier = 3
        else:
            tier = 2  # default to medium

        travel_tier[(from_team, to_team)] = tier
```

```python
# sort by team and date
df = df.sort_values(by=['team', 'date'])

# get previous game's opponent and home/away
df['prev_opp'] = df.groupby('team')['opp'].shift(1)
df['prev_home_away'] = df.groupby('team')['home_away'].shift(1)
```

```python
# define where the team played based on home_away
def get_city(team, opp, home_away):
    return team if home_away == 1 else opp
```

```python
# apply to current and previous game
df['curr_city'] = df.apply(lambda row: get_city(row['team'], row['opp'], row['home_away']), axis=1)
df['prev_city'] = df.apply(lambda row: get_city(row['team'], row['prev_opp'], row['prev_home_away']), axis=1)
```

```python
# update travel_distance using new travel_tier map
df['travel_distance'] = df.apply(
    lambda row: travel_tier.get((row['prev_city'], row['curr_city']), None),
    axis=1
)
```

```python
def add_grouped_medians(df, columns, group_keys, suffix):
    # compute medians
    grouped = df.groupby(group_keys)[columns].median()

    # rename columns
    rename_map = {col: f"{col}_by_{'_'.join(group_keys)}" for col in columns}
    grouped = grouped.rename(columns=rename_map)

    # reset index after renaming (avoids clash)
    grouped = grouped.reset_index()

    # only keep new columns + group keys
    new_cols = group_keys + [col for col in grouped.columns if col not in df.columns and col not in group_keys]
    grouped = grouped[new_cols]

    # merge back
    df = df.merge(grouped, on=group_keys, how='left')
    return df
```

```python
# all numeric columns except target and ID stuff
cols_to_expand = [col for col in df.columns
                  if df[col].dtype != 'object'
                  and col not in ['team_score', 'opp_score', 'g_num']]

# groupings to try
df = add_grouped_medians(df, cols_to_expand, ['team', 'home_away'], 'by_home_away')
df = add_grouped_medians(df, cols_to_expand, ['team', 'day_of_week'], 'by_day')
df = add_grouped_medians(df, cols_to_expand, ['team', 'travel_distance'], 'by_travel')
```
