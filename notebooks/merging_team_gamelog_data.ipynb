```python
import pandas as pd
```

```python
# load the player game logs CSV from the data folder
df = pd.read_csv("LHL-final-final-project/data/2024_basketball_reference_gamelog.csv")
df_adv = pd.read_csv("LHL-final-final-project/data/2024_basketball_reference_gamelog-advanced.csv")

# preview
df.head()
df_adv.head()
```

```python
# define merge keys
merge_keys = ['team', 'year', 'month', 'day']

# find columns in df_adv that are not already in df (excluding merge keys)
new_cols = [col for col in df_adv.columns if col not in df.columns or col in merge_keys]

# perform the merge
df = df.merge(df_adv[new_cols], on=merge_keys, how='left')
```
