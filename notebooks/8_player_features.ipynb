```python
# create a 2024-only copy if you haven't already
df_2024_only = df[df['year'] == 2024].copy()

# rank each offensive stat within team (lower rank = better)
df_2024_only['usg_rank'] = df_2024_only.groupby('tm')['advanced_usg_pct'].rank(ascending=False)
df_2024_only['ortg_rank'] = df_2024_only.groupby('tm')['per_poss_ortg'].rank(ascending=False)
df_2024_only['ows_rank'] = df_2024_only.groupby('tm')['advanced_ows'].rank(ascending=False)
```

```python
df_2024_only
```

```python
# sum the three ranks to get a total offensive rank score (lower = better)
df_2024_only['off_rank_total'] = (
    df_2024_only['usg_rank'] +
    df_2024_only['ortg_rank'] +
    df_2024_only['ows_rank']
)
```

```python
df_2024_only
```

```python
# filter to players with at least 18 minutes per game
eligible = df_2024_only[df_2024_only['per_game_mp'] >= 18]

# get top 3 offensive players per team based on total rank
top3_offensive = (
    eligible.sort_values(['tm', 'off_rank_total'])
            .groupby('tm')
            .head(3)
)

# view selected columns
top3_offensive[['player', 'tm', 'per_game_mp', 'off_rank_total', 'advanced_usg_pct', 'per_poss_ortg', 'advanced_ows']]
```

```python
# rank defensive stats within team (2024 only)
df_2024_only['drtg_rank'] = df_2024_only.groupby('tm')['per_poss_drtg'].rank(ascending=True)  # lower is better
df_2024_only['dws_rank'] = df_2024_only.groupby('tm')['advanced_dws'].rank(ascending=False)   # higher is better
df_2024_only['usg_def_rank'] = df_2024_only.groupby('tm')['advanced_usg_pct'].rank(ascending=True)  # lower = more defensive/off-ball
```

```python
df_2024_only
```

```python
# sum the defensive ranks (lower total = better overall defensive profile)
df_2024_only['def_rank_total'] = (
    df_2024_only['drtg_rank'] +
    df_2024_only['dws_rank'] +
    df_2024_only['usg_def_rank']
)
```

```python
df_2024_only
```

```python
# filter to players with at least 18 minutes per game
eligible_def = df_2024_only[df_2024_only['per_game_mp'] >= 18]

# get top 3 defensive players per team based on total defensive rank
top3_defensive = (
    eligible_def.sort_values(['tm', 'def_rank_total'])
                .groupby('tm')
                .head(3)
)

# view selected columns
top3_defensive[['player', 'tm', 'per_game_mp', 'def_rank_total', 'per_poss_drtg', 'advanced_dws', 'advanced_usg_pct']]
```

```python
# get just the player names
offensive_players = set(top3_offensive['player'])
defensive_players = set(top3_defensive['player'])

# players in both
both = offensive_players & defensive_players

# offensive-only
off_only = offensive_players - defensive_players

# defensive-only
def_only = defensive_players - offensive_players

# show results
print(f"Players in both: {len(both)}")
print(f"Offensive only: {len(off_only)}")
print(f"Defensive only: {len(def_only)}")

# optional: see names
print("\nPlayers in both:")
print(both)

print("\nOffensive only:")
print(off_only)

print("\nDefensive only:")
print(def_only)
```

```python
# remove "TOT" rows before final team-based selections
df_2024_only = df_2024_only[df_2024_only['tm'] != 'TOT']
```

```python
# get list of unique player-team combos we want to keep
players_to_keep = pd.concat([top3_offensive, top3_defensive], ignore_index=True)[['player', 'tm']].drop_duplicates()

# merge with full 2024 data to retain all columns
combined = df_2024_only.merge(players_to_keep, on=['player', 'tm'], how='inner')

# optional: sort for readability
combined = combined.sort_values(['tm', 'player'])
```

```python
combined
```

```python
# players from combined
combined_players = set(combined['player'])

# players from offense + defense lists
expected_players = set(top3_offensive['player']) | set(top3_defensive['player'])

# compare
missing_from_combined = expected_players - combined_players
extra_in_combined = combined_players - expected_players

print(f"Players in offense/defense lists but missing from combined: {missing_from_combined}")
print(f"Players in combined but not in offense/defense lists: {extra_in_combined}")
```

```python
# drop TOT rows
combined = combined[combined['tm'] != 'TOT'].copy()

# drop temp ranking and helper columns
combined = combined.drop(columns=[
    'weighted_ortg',
    'usg_rank', 'ortg_rank', 'ows_rank', 'off_rank_total',
    'drtg_rank', 'dws_rank', 'usg_def_rank', 'def_rank_total'
], errors='ignore')

# optional: re-sort
combined = combined.sort_values(['tm', 'player'])

# display result
combined.head()
```

```python
# create a key column in both DataFrames
df_model['game_key'] = df_model['team'] + "_" + df_model['opp'] + "_" + df_model['month'].astype(str) + "_" + df_model['day'].astype(str)
player_matrix['game_key'] = player_matrix['tm'] + "_" + player_matrix['opp'] + "_" + player_matrix['month'].astype(str) + "_" + player_matrix['day'].astype(str)

# find missing games
missing_games = df_model[~df_model['game_key'].isin(player_matrix['game_key'])]

# display count and rows
print(f"Missing games: {missing_games.shape[0]}")
missing_games[['team', 'opp', 'month', 'day']]
```

```python
# identify player columns only (exclude game identifiers)
player_columns = [col for col in player_matrix.columns if col not in ['tm', 'opp', 'month', 'day', 'game_key']]

# merge player participation into df_model
df_model_merged = df_model.merge(
    player_matrix[['tm', 'opp', 'month', 'day'] + player_columns],
    left_on=['team', 'opp', 'month', 'day'],
    right_on=['tm', 'opp', 'month', 'day'],
    how='left'
).drop(columns=['tm'])
```

```python
df_model_merged.shape
```

```python
df_model_merged[player_columns] = df_model_merged[player_columns].fillna(0).astype(int)
```
